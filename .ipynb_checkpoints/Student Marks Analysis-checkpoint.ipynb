```python
#gender Distribution
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x='Gender')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
# From above chart we have analyzed that:
# The number of females in the data is more than the number of males.
```

```python
gp= df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gp)
```

```python
plt.figure(figsize=(5,3))
sns.heatmap(gp,annot=True)
plt.show()
```

```python
# From the above chart we have concluded that the education of the parents have good impact on their scores.
```

```python
print(df["EthnicGroup"].unique())
```

```python
# Distribution of Ethnic Group
grpA=df.loc[(df["EthnicGroup"]=="group A")].count()
grpB=df.loc[(df["EthnicGroup"]=="group B")].count()
grpC=df.loc[(df["EthnicGroup"]=="group C")].count()
grpD=df.loc[(df["EthnicGroup"]=="group D")].count()
grpE=df.loc[(df["EthnicGroup"]=="group E")].count()

l=["grpA","grpB","grpC","grpD","grpE"]
mylist=[grpA["EthnicGroup"],grpB["EthnicGroup"],grpC["EthnicGroup"],grpD["EthnicGroup"],grpE["EthnicGroup"]]

plt.pie(mylist,labels=l,autopct="%1.2f%%")
plt.show()
```
